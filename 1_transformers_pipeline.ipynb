```python
from transformers import pipeline
classifier = pipeline("sentiment-analysis")
```

```python
samp1=classifier("I've been waiting for a HuggingFace course my whole life.")
samp2=classifier("The beginning of the movie was interesting, but when the hero died, he became naive")
print(samp1,samp2)
```
